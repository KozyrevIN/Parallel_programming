```python
import pandas as pd
import matplotlib.pyplot as plt

plt.xlabel('threads')
plt.ylabel('time, s')

res = pd.read_csv('out.csv')

plt.plot(res['n'], res['t'])

plt.show()
```

```python
plt.xlabel('threads')
plt.ylabel('boost')

res = pd.read_csv('out.csv')

plt.plot(res['n'], res['t'][0] / res['t'])

plt.show()
```

```python
plt.xlabel('threads')
plt.ylabel('efficency')

res = pd.read_csv('out.csv')

plt.plot(res['n'], res['t'][0] / (res['t'] * res['n']))

plt.show()
```
